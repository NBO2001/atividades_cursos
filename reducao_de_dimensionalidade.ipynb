# Lendo e monstando imagem colorida

```python
img = cv2.imread("./lena.png")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis('off')
plt.show()
```

# Gerando uma imagem em tons de cinza

```python
img = img[:,:,0] * 0.21 + img[:,:,0] * 0.72 + img[:,:,0] * 0.07
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

# Gerando a imagem binarizadad

```python
limiar = 162.5

h,w = img.shape

for x in range(h):
  for y in range(w):
    if img[x,y] >= limiar:
      img[x,y] = 1
    else:
      img[x,y] = 0


plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```
